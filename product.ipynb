```python
#Top Categories
df['category'].value_counts().head(5)
```

```python
import matplotlib.pyplot as plt

df['category'].value_counts().head(10).plot(kind='bar')

plt.title("Top 10 Product Categories")
plt.xlabel("Category")
plt.ylabel("Count")

plt.show()
```

```python
#Insights 
top_category = df['category'].value_counts().idxmax()

print("Most Popular Category:", top_category)
```

```python
print(df['category'].value_counts().head(5))
```

```python
category_percent = df['category'].value_counts(normalize=True) * 100

print(category_percent.head(10))
```

```python
#Observations

#Insights:
#1. LaptopSpeaker is the most popular category.
#2. MobileSpeaker and AudioMP3Player also have high counts.
#3. Some categories have very low frequency.
#4. Dataset is highly imbalanced.
```

```python
df.to_csv("cleaned_products.csv", index=False)

print("Dataset Saved Successfully")
```

```python
#Pie Chart for Percentage Distribution

df['category'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',
    figsize=(8,8)
)

plt.title("Category Distribution")
plt.ylabel("")

plt.show()
```

```python
#Countplot using Seaborn

import seaborn as sns

plt.figure(figsize=(10,5))

sns.countplot(
    y=df['category'],
    order=df['category'].value_counts().index
)

plt.title("Category Countplot")

plt.show()
```

```python
df['super_category'].value_counts().plot(kind='bar')

plt.title("Super Category Distribution")
plt.xlabel("Super Category")
plt.ylabel("Count")

plt.show()
```

```python
plt.figure(figsize=(12,6))

df['sub_category'].value_counts().head(10).plot(kind='bar')

plt.title("Top 10 Sub Categories")
plt.xlabel("Sub Category")
plt.ylabel("Count")

plt.xticks(rotation=45)

plt.show()
```

```python
plt.figure(figsize=(12,6))

df['vertical'].value_counts().head(10).plot(kind='bar')

plt.title("Top 10 Verticals")
plt.xlabel("Vertical")
plt.ylabel("Count")

plt.xticks(rotation=45)

plt.show()
```
